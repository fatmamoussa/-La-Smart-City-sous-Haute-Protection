```python
# =============================================================================
# BLOC 2: GÃ‰NÃ‰RATEUR DE DONNÃ‰ES RÃ‰ALISTES (SFAX)
# =============================================================================
# Simule les capteurs IoT de la ville

class GenerateurDonneesSfax:
    """
    GÃ©nÃ¨re des donnÃ©es rÃ©alistes pour la ville de Sfax
    - Patterns journaliers (pics matin/soir)
    - Variations alÃ©atoires rÃ©alistes
    - Attaques FDI alÃ©atoires
    """

    def __init__(self):
        self.temps = 0
        print(" Initialisation du gÃ©nÃ©rateur Sfax...")

    def generer(self):
        """GÃ©nÃ¨re une nouvelle donnÃ©e"""
        heure = self.temps % 24

        # ===== Ã‰LECTRICITÃ‰ (Ã©clairage public) =====
        if 18 <= heure <= 22:  # Soir - pic d'Ã©clairage
            elec = 850 + 50 * np.sin(self.temps * 0.3) + np.random.normal(0, 25)
        elif 7 <= heure <= 9:   # Matin
            elec = 620 + 40 * np.cos(self.temps * 0.2) + np.random.normal(0, 20)
        else:                    # Autres heures
            elec = 380 + 30 * np.sin(self.temps * 0.1) + np.random.normal(0, 15)

        # ===== EAU (distribution) =====
        if 7 <= heure <= 9:      # Pic matinal
            eau = 3800 + 250 * np.sin(self.temps * 0.4) + np.random.normal(0, 80)
        elif 12 <= heure <= 14:  # Pic midi
            eau = 4200 + 200 * np.cos(self.temps * 0.3) + np.random.normal(0, 90)
        elif 19 <= heure <= 21:  # Pic soir
            eau = 4000 + 180 * np.sin(self.temps * 0.5) + np.random.normal(0, 70)
        else:
            eau = 2500 + 150 * np.cos(self.temps * 0.2) + np.random.normal(0, 60)

        # ===== ATTAQUES FDI (8% des donnÃ©es) =====
        est_attaque = np.random.random() < 0.08
        type_attaque = None

        if est_attaque:
            type_attaque = np.random.choice(['SURGE', 'DROP', 'STEALTH'])

            if type_attaque == 'SURGE':  # Pic artificiel (risque blackout)
                elec *= np.random.uniform(2.5, 3.8)
            elif type_attaque == 'DROP':  # Chute brutale (cache fuite)
                eau *= np.random.uniform(0.18, 0.32)
            else:  # STEALTH - attaque furtive
                elec *= 0.45
                eau *= 0.45

        donnee = {
            'temps': self.temps,
            'heure': heure,
            'elec': max(elec, 80),
            'eau': max(eau, 400),
            'est_attaque': est_attaque,
            'type_attaque': type_attaque
        }

        self.temps += 1
        return donnee

# Test rapide
gen_test = GenerateurDonneesSfax()
test = gen_test.generer()
print(f"\n Test gÃ©nÃ©rateur: Ã‰lec={test['elec']:.0f} kWh, Eau={test['eau']:.0f} L/h")
print("="*70)
```

```python
# =============================================================================
# BLOC 3: DÃ‰TECTEUR ISOLATION FOREST (IA)
# =============================================================================
# ModÃ¨le non-supervisÃ© pour dÃ©tecter les anomalies

class DetecteurIA:
    """
    Isolation Forest - DÃ©tecte les attaques sans apprentissage supervisÃ©
    """

    def __init__(self):
        self.model_elec = IsolationForest(contamination=0.08, random_state=42)
        self.model_eau = IsolationForest(contamination=0.08, random_state=42)
        self.scaler = StandardScaler()
        self.entraine = False

    def entrainer(self, df_entrainement):
        """EntraÃ®ne les modÃ¨les sur donnÃ©es normales"""
        print("\n ENTRAÃNEMENT ISOLATION FOREST")
        print("-" * 40)

        if len(df_entrainement) == 0:
            print(" Erreur: Pas de donnÃ©es")
            return

        # Features: consommation + heure (cycle journalier)
        X_elec = df_entrainement[['elec', 'heure']].values
        X_eau = df_entrainement[['eau', 'heure']].values

        # Normalisation
        X_elec_scaled = self.scaler.fit_transform(X_elec)
        X_eau_scaled = self.scaler.fit_transform(X_eau)

        # EntraÃ®nement
        print("   â€¢ ModÃ¨le Ã©lectricitÃ©...")
        self.model_elec.fit(X_elec_scaled)

        print("   â€¢ ModÃ¨le eau...")
        self.model_eau.fit(X_eau_scaled)

        self.entraine = True

        # Ã‰valuation rapide
        pred_elec = self.model_elec.predict(X_elec_scaled)
        pred_eau = self.model_eau.predict(X_eau_scaled)

        anomalies_elec = sum(pred_elec == -1)
        anomalies_eau = sum(pred_eau == -1)

        print(f"\n ModÃ¨les entraÃ®nÃ©s")
        print(f"   â€¢ Anomalies dÃ©tectÃ©es (entraÃ®nement):")
        print(f"     - Ã‰lectricitÃ©: {anomalies_elec} points")
        print(f"     - Eau: {anomalies_eau} points")
        print("="*70)

    def detecter(self, elec, eau, heure):
        """DÃ©tecte si c'est une attaque en temps rÃ©el"""
        if not self.entraine:
            return False, 0.5

        # PrÃ©paration
        X_elec = np.array([[elec, heure]])
        X_eau = np.array([[eau, heure]])

        # Normalisation
        X_elec_scaled = self.scaler.transform(X_elec)
        X_eau_scaled = self.scaler.transform(X_eau)

        # PrÃ©diction (-1 = anomalie)
        pred_elec = self.model_elec.predict(X_elec_scaled)[0]
        pred_eau = self.model_eau.predict(X_eau_scaled)[0]

        # Score de confiance
        score_elec = self.model_elec.score_samples(X_elec_scaled)[0]
        score_eau = self.model_eau.score_samples(X_eau_scaled)[0]

        est_attaque = (pred_elec == -1) or (pred_eau == -1)
        confiance = min(1 - (score_elec + score_eau) / (-2), 0.99)

        return est_attaque, confiance

print("\n BLOC 3: DÃ‰TECTEUR PRÃŠT")
print("="*70)
```

```python
# =============================================================================
# BLOC 4: SIMULATEUR TEMPS RÃ‰EL
# =============================================================================
# GÃ¨re le flux de donnÃ©es et l'optimisation

class SimulateurTempsReel:
    """
    Simule le flux continu de donnÃ©es de la ville
    """

    def __init__(self, detecteur):
        self.generateur = GenerateurDonneesSfax()
        self.detecteur = detecteur
        self.historique = []
        self.alertes = []

    def optimiser_elec(self, valeur, heure):
        """Optimisation intelligente de l'Ã©lectricitÃ©"""
        if heure >= 23 or heure < 6:      # Nuit - rÃ©duction max
            return valeur * 0.55
        elif 10 <= heure <= 15:            # Mi-journÃ©e
            return valeur * 0.75
        elif 18 <= heure <= 22:            # Soir (pic)
            return valeur * 0.80
        return valeur * 0.70

    def optimiser_eau(self, valeur, heure):
        """Optimisation intelligente de l'eau"""
        if 10 <= heure <= 16:              # Heures creuses
            return valeur * 0.65
        elif 7 <= heure <= 9 or 19 <= heure <= 21:  # Heures de pointe
            return valeur * 0.80
        return valeur * 0.75

    def prochaine_donnee(self):
        """GÃ©nÃ¨re et traite la prochaine donnÃ©e"""
        donnee = self.generateur.generer()

        # DÃ©tection IA
        est_attaque, confiance = self.detecteur.detecter(
            donnee['elec'], donnee['eau'], donnee['heure']
        )

        # Si attaque dÃ©tectÃ©e, ajouter aux alertes
        if est_attaque and donnee['est_attaque']:
            self.alertes.append({
                'temps': donnee['temps'],
                'type': donnee['type_attaque'],
                'elec': donnee['elec'],
                'eau': donnee['eau'],
                'confiance': confiance
            })
            # Garder 5 derniÃ¨res alertes
            if len(self.alertes) > 5:
                self.alertes.pop(0)

        # Optimisation
        if est_attaque:
            elec_opt = self.optimiser_elec(donnee['elec'], donnee['heure'])
            eau_opt = self.optimiser_eau(donnee['eau'], donnee['heure'])
        else:
            elec_opt = donnee['elec']
            eau_opt = donnee['eau']

        # Sauvegarde historique
        self.historique.append({
            'temps': donnee['temps'],
            'elec_reelle': donnee['elec'],
            'elec_optimise': elec_opt,
            'eau_reelle': donnee['eau'],
            'eau_optimise': eau_opt,
            'est_attaque': donnee['est_attaque'],
            'type_attaque': donnee['type_attaque']
        })

        # Garder 100 points
        if len(self.historique) > 100:
            self.historique.pop(0)

        return donnee

print("\n BLOC 4: SIMULATEUR PRÃŠT")
print("="*70)
```
